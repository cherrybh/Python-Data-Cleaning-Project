```python
def compare_address(row):
    if fuzz.token_sort_ratio(row["FullAddress"], row["OwnerStAddress"]) > 50:
        return 'Matches'
    else:
        return row['FullAddress']

par['First_Comparison'] = par.apply(lambda row: compare_address(row), axis=1)
```

```python
import pandas as pd 
```

```python
df = pd.read_excel("ParcelSub.xlsx")
```

```python
df.head(100)
```

```python
df['Zip'] = df['Zip'].fillna(0).astype(int)
```

```python
df.dtypes
```

```python
df['Zip'].replace(['0', '0.0'], '', inplace=True)
```

```python
df['fulladd_assess'] = df[["FullAddress", "Zip"]].astype(str).apply(lambda x: " SYRACUSE, NY ".join(x), axis=1)
```

```python
df.head()
```

```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="aybhatna@syr.edu")
```

```python
def extract_clean_address(address):
    try:
        location = geolocator.geocode(address)
        return location.address
    except:
        return ''
df['cleaned_address'] = df.apply(lambda x: extract_clean_address(x['fulladd_assess']) , axis =1  )
```
